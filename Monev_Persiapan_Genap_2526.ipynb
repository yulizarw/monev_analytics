```python
kepatuhan_prodi = (
    df.groupby('namaunit')
      .apply(lambda x: (1 - x.isnull().mean()).mean() * 100)
      .reset_index(name='tingkat_kepatuhan')
)

kepatuhan_prodi.sort_values('tingkat_kepatuhan', ascending=False).head()
```

```python
kepatuhan_prodi.sort_values('tingkat_kepatuhan').plot(
    x='namaunit', y='tingkat_kepatuhan',
    kind='barh', figsize=(8,10)
)
plt.title("Tingkat Pengisian Monev per Prodi")
plt.xlabel("Persentase Pengisian(%)")
plt.ylabel("Program Studi")
plt.show()
```

```python
konsistensi = (
    df.groupby(['namaunit', 'perioderm'])
      .size()
      .reset_index(name='jumlah_data')
)

konsistensi.head()
stabilitas = (
    konsistensi.groupby('namaunit')['jumlah_data']
    .std()
    .reset_index(name='deviasi_pelaporan')
    .sort_values('deviasi_pelaporan', ascending=False)
)

stabilitas.head()
early_warning = kepatuhan_prodi.merge(
    stabilitas, on='namaunit', how='left'
)

early_warning['status_risiko'] = np.where(
    (early_warning['tingkat_kepatuhan'] < 70) |
    (early_warning['deviasi_pelaporan'] > early_warning['deviasi_pelaporan'].median()),
    'PERLU INTERVENSI',
    'TERKENDALI'
)

early_warning.sort_values('tingkat_kepatuhan').head(10)
```

```python
prodi_kritis = early_warning[early_warning['status_risiko']=='PERLU INTERVENSI']

print(f"""
Hasil analisis awal monev pembelajaran menunjukkan bahwa
dari total {df['namaunit'].nunique()} program studi,
sebanyak {len(prodi_kritis)} prodi teridentifikasi
memerlukan perhatian khusus.

Indikator utama yang digunakan meliputi tingkat kepatuhan
pengisian instrumen monev serta konsistensi pelaporan
antar periode. Analisis ini bersifat sementara dan
digunakan sebagai bentuk kendali mutu awal.
""")
```

```python
fe = early_warning.copy()

# Normalisasi sederhana
fe['kepatuhan_norm'] = fe['tingkat_kepatuhan'] / 100
fe['stabilitas_norm'] = 1 - (
    fe['deviasi_pelaporan'] / fe['deviasi_pelaporan'].max()
)

# Skor mutu komposit
fe['indeks_mutu_prodi'] = (
    0.5 * fe['kepatuhan_norm'] +
    0.5 * fe['stabilitas_norm']
)

fe[['namaunit', 'indeks_mutu_prodi']].sort_values(
    'indeks_mutu_prodi', ascending=False
).head()

fe['level_mutu'] = pd.cut(
    fe['indeks_mutu_prodi'],
    bins=[0, 0.6, 0.8, 1],
    labels=['MERAH', 'KUNING', 'HIJAU']
)

fe['level_mutu'].value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import re
```

```python
stopwords_id = set([
    # kata hubung umum
    "dan","atau","yang","pada","dalam","dengan","untuk","dari","ke","sebagai",
    "oleh","terhadap","tentang","serta","melalui","agar","guna","adalah",

    # kata administratif / generik monev
    "form","data","dokumen","laporan","pengisian","isian",
    "nilai","persen","prosentase","realisasi",
    "monev","monitoring","evaluasi","pembelajaran",
    "program","studi","prodi","unit",
    "tahun","periode","semester",

    # kata umum tidak informatif
    "kegiatan","pelaksanaan","pelaksana","hasil","target","standar"
])

stopwords = STOPWORDS.union(stopwords_id)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)  # hapus angka & simbol
    text = re.sub(r'\s+', ' ', text)          # rapikan spasi
    return text
```

```python
text_target = " ".join(
    df['namatarget']
    .dropna()
    .astype(str)
    .apply(clean_text)
)

wordcloud_target = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    stopwords=stopwords,
    min_word_length=4,
    collocations=False
).generate(text_target)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud_target, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud Informatif Target Mutu Monev Pembelajaran", fontsize=14)
plt.show()
```

```python
text_target_kosong = " ".join(
    df[df['realisasi'].isnull()]['namatarget']
    .dropna()
    .astype(str)
    .apply(clean_text)
)

wordcloud_kosong = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    stopwords=stopwords,
    colormap='Reds',
    min_word_length=4,
    collocations=False
).generate(text_target_kosong)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud_kosong, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud Informatif Target Mutu dengan Realisasi Kosong", fontsize=14)
plt.show()
```
